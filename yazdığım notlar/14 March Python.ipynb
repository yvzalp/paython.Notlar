```python
def slicer(*args) :    # benim zm
    odd_list = []
    even_list = []
    for i in args :
        if i % 2 :
            odd_list.append(i)
        else :
            even_list.append(i)
    print("even list", even_list)
    print("odd list", odd_list)
slicer(1,2,3,4,5,6,7,8,9)
```

```python
def slicer(*num) :
    
    evens = []
    odds = []
    
    for i in num :
        if i % 2 == 0 :
            evens.append(i)
        else :
            odds.append(i)
    print("Even list : ", evens)
    print("Odd list : ", odds)
    
```

```python
slicer(1,2,3,4,5,6)
```

- **kwargs kullanlarak tanmlanan parametre dict muamelesi grr. arrken de keywordslerle arlmas lazm

```python
def animals(**kwargs) :
    for key, value in kwargs.items() :
        print(value, "are", key)
        
animals(Carnivores="Lions", Omnivores="Bears", Herbivores="Deers", Nomnivores="Human")
```

```python
def organizer(**kwargs) :   # bu ilk zmm
    print(kwargs.values())
    print(kwargs.keys())
organizer(Beth=26, Oscar=42, Justin=18, Frank=33)
```

- filter()  => Construct an iterator from those elements of iterable for which function returns true. iterable may be either a sequence, a container which supports iteration, or an iterator. If function is None, the identity function is assumed, that is, all elements of iterable that are false are removed.
- filter() metodu: iterable bir eyi ilemenize ve belirli bir koulu karlayan eleri karmanza (extract etmeye) olanak tanyan build in fonksiyondur

```python
cumle = "Bu bir cmledir. Sesli harfleri seilecektir."
def secme(b):    # benim zm
    
    a = "a e  i u  o ".split()
    
    if b.lower() in a :
        return True
    else:
        return False
    
c = filter(secme, cumle)
print(*c)
```

```python
cmle = "ben size ders anlatrkene HUZUR buluyorum. dertleri unutuyorum" # hocann zm.
def vowel(letter) :
    
    vowels = "a e i u o   ".split()
    
    if letter.lower() in vowels :
        return True
    else : 
        return False
    
```

```python
filtered_vowels = filter(vowel, cmle)  # fonksiyonu sra sra tatbik eder ve istenileni seer.
```

```python
print(* filtered_vowels)
```
