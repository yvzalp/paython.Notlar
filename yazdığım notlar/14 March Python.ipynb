- Fonksiyonu çağırırken fonksiyona birden çok argüman yollayabiliyoruz.
- Bunu yapabilmek için fonksiyonda parametrenin başına * koyuyoruz.
- Önüne * koyduğumuzda o  parametre iterable hale gelmiş oluyor. Bu nedenle birden fazla değer alabiliyor.

```python
def fruiterer(*fruit) : # sayısı belli olmayan kadar arguman çağrılabilir. * iterable yapıyor.
    print("I want to get : ")
    for i in fruit :
        print("-", i)
fruiterer("orange", "banana", "melon", "ananas")
```

```python
def fruit(*fruitt) :   # tuple çıktı verir.
    print(type(fruitt))
fruit("banana")
```

- tipi tuple çünkü fruit'e yolladığımız argüman şudur: 'orange', 'banana', 'melon', 'ananas'
- Hatırlayalım; arasına virgül koyarak PARANTEZSİZ yazdığımız objeler tuple oluyordu

```python
def slicer(*args) :    # benim çözüm
    odd_list = []
    even_list = []
    for i in args :
        if i % 2 :
            odd_list.append(i)
        else :
            even_list.append(i)
    print("even list", even_list)
    print("odd list", odd_list)
slicer(1,2,3,4,5,6,7,8,9)
```

```python
def slicer(*num) :
    
    evens = []
    odds = []
    
    for i in num :
        if i % 2 == 0 :
            evens.append(i)
        else :
            odds.append(i)
    print("Even list : ", evens)
    print("Odd list : ", odds)
    
```

```python
slicer(1,2,3,4,5,6)
```

- **kwargs kullanılarak tanımlanan parametre dict muamelesi görür. Çağırırken de keywordslerle çağrılması lazım

```python
def animals(**kwargs) :
    for key, value in kwargs.items() :
        print(value, "are", key)
        
animals(Carnivores="Lions", Omnivores="Bears", Herbivores="Deers", Nomnivores="Human")
```

```python
def organizer(**kwargs) :   # bu ilk çözümüm
    print(kwargs.values())
    print(kwargs.keys())
organizer(Beth=26, Oscar=42, Justin=18, Frank=33)
```

```python
def organizer(**kwargs) :   # bu ikinci çözümüm
    isimler = []
    yaslar = []
    for key, values in kwargs.items() :
        isimler.append(key)
        yaslar.append(values)
    print(isimler)
    print(yaslar)
organizer(Beth=26, Oscar=42, Justin=18, Frank=33)
```

```python
def organizer(** people) :  # Joseph hocanın çözümü
    
    names = []
    ages = []
    
    for keys, values in people.items() :
        names.append(keys)
        ages.append(values)
    print(names, ages, sep = "\n")
```

```python
organizer(Beth = 26, Oscar = 41, Justin = 18)
```

```python
organizer(Beth = 26, Oscar = 41, Justin = 18, Frank = 33)
```

- filter()  => Construct an iterator from those elements of iterable for which function returns true. iterable may be either a sequence, a container which supports iteration, or an iterator. If function is None, the identity function is assumed, that is, all elements of iterable that are false are removed.
- filter() metodu: iterable bir öğeyi işlemenize ve belirli bir koşulu karşılayan öğeleri çıkarmanıza (extract etmeye) olanak tanıyan build in fonksiyondur

```python
cumle = "Bu bir cümledir. Sesli harfleri seçilecektir."
def secme(b):    # benim çözüm
    
    a = "a e ı i u ü o ö".split()
    
    if b.lower() in a :
        return True
    else:
        return False
    
c = filter(secme, cumle)
print(*c)
```

```python
cümle = "ben size ders anlatırkene HUZUR buluyorum. dertleri unutuyorum" # hocanın çözümü.
def vowel(letter) :
    
    vowels = "a e i u o ö ı ü".split()
    
    if letter.lower() in vowels :
        return True
    else : 
        return False
    
```

```python
filtered_vowels = filter(vowel, cümle)  # fonksiyonu sıra sıra tatbik eder ve istenileni seçer.
```

```python
print(* filtered_vowels)
```
